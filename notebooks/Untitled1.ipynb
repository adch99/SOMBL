```python
%cd /home/aditya/Documents/SOMBL/
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
r_reg1 = np.linspace(0, 1, 100)
r_reg2 = np.linspace(1, 5, 400)
r_reg3 = np.linspace(5, 7, 200)

slope = -0.1

y_reg1 = slope*r_reg2[0] + 0.03*r_reg1[-1] - 0.03*r_reg1 + 0.01 * np.random.normal(size=(100))
y_reg2 = slope*r_reg2
y_reg3 = y_reg2[-1] - 0.01*(r_reg3 - r_reg2[-1]) + 0.01 * np.random.normal(size=(200))

r = np.empty(700)
y = np.empty(700)

r[:100] = r_reg1
r[100:500] = r_reg2
r[500:700] = r_reg3

y[:100] = y_reg1
y[100:500] = y_reg2
y[500:700] = y_reg3

y -= y.min()

with plt.xkcd():
    # plt.plot(r_reg1, y_reg1, color="blue")
    # plt.plot(r_reg2, y_reg2, color="blue")
    # plt.plot(r_reg3, y_reg3, color="blue")
    plt.plot(r, y)
    plt.plot(np.ones(100)*5, np.linspace(y.min(), y.max(), 100), linestyle="--")
    plt.xticks([5], labels=["L"])
    plt.yticks([])
    plt.ylabel(r"$\log\left\langle|G^R_{\alpha\beta}|^2\right\rangle_{dis}(r)$", size="xx-large")
    plt.xlabel(r"$r$", size="xx-large")
    ax = plt.gca()
    ax.spines[["left", "bottom"]].set_position(("data", 0))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False, zorder=3)
    plt.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False, zorder=3)

fname = "gfuncsq_vs_r_schematic"
plt.savefig("plots/PDFs/" + fname + ".pdf", bbox_inches="tight")
plt.savefig("plots/PNGs/" + fname + ".png", bbox_inches="tight")
```
