```python
M = np.random.normal(size=(10, 10))
A = np.abs(M)**2
S1 = np.dot(A, A.T)

S2 = np.zeros(M.shape)
for i in range(10):
    for j in range(10):
        for n in range(10):
            S2[i, j] += np.abs(M[j, n])**2 * np.abs(M[i, n])**2

diffS = S2 - S1
print(f"Diff: {np.sum(np.abs(diffS))}")

plt.matshow(diffS)
plt.colorbar()
```

```python
M = np.random.normal(loc=100, scale=1000, size=(10, 10)) + 1j*np.random.normal(size=(10,10))
X = np.conj(M[:, ::2]) * M[:, 1::2]
B = np.zeros((10,5), dtype=complex)
for i in range(10):
    for n in range(5):
        B[i,n] = np.conj(M[i,2*n]) * M[i,2*n+1]

print(np.sum(np.abs(B-X)))
BBh = B @ B.conj().T
S1 = BBh + np.conj(BBh)

S2 = np.zeros(M.shape, dtype=complex)
for i in range(10):
    for j in range(10):
        for n in range(5):
            term = np.conj(M[i,2*n]) * M[j,2*n] * np.conj(M[j,2*n+1]) * M[i,2*n+1]
            S2[i,j] += term + np.conj(term)
            
diffS = S2 - S1
print(f"Diff: {np.sum(np.abs(diffS))}")

plt.matshow(np.abs(diffS))
plt.colorbar()
```
